```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv") #median_house_value is our target variable
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace = True) #making every variable have the same number of non-null values
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis = 1)
y = data['median_house_value'] #generated independent and dependent data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
train_data = x_train.join(y_train) #all the x and y training data
```

```python
train_data
```

```python
train_data.hist(figsize = (15,8))
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu") #heatmap of the correlation matrix between the values(each value has a corr of 1 with itself)
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] +1) #taking the log to make a normal distribution
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] +1)
train_data['population'] = np.log(train_data['population'] +1)
train_data['households'] = np.log(train_data['households'] +1)
```

```python
train_data.hist()
```

```python
train_data.ocean_proximity.value_counts() #we don't want categorical data 
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1) #creating dummy variables to turn categorical into binary data
```

```python
train_data
```

```python
plt.figure(figsize = (15, 8))
sns.scatterplot(x='latitude', y='longitude', data = train_data, hue = 'median_house_value', palette = 'coolwarm') #making a scatterplot of coordinates, redder = more expensive, blue = less expensive
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms'] #feature engineering for room ratios, which are more interesting to analyze
train_data['household_rooms'] = train_data['total_rooms']/train_data['households'] #work on feature engineering more features in the future
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression #creating a linear regression


x_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)
```

```python
test_data['total_rooms'] = np.log(test_data['total_rooms'] +1) #taking the log to make a normal distribution
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] +1)
test_data['population'] = np.log(test_data['population'] +1)
test_data['households'] = np.log(test_data['households'] +1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)

test_data['bedroom_ratio'] = test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms']/test_data['households']
```

```python
x_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
reg.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor #using a random forest regression model for more accuracy

forest = RandomForestRegressor()

```

```python
from sklearn.model_selection import GridSearchCV #using grid search cross-validation for more accurate results
# Define the parameter grid for hyperparameter tuning
parameter_grid = {
    "n_estimators": [100,200,300],
    "min_samples_split": [2,4],
    "max_depth": [None, 4, 8]
}

grid_search = GridSearchCV(forest, parameter_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)

# Fit the model using the grid search on the training data
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(x_test, y_test)
```
